1、 导入数据+查看原始数据信息

```python
import mne
from mne.datasets.opm import data_path

data_path = r'data/BCICIV_2a_gdf/A01T.gdf'
raw = mne.io.read_raw_gdf(data_path, preload=True)

print(raw)
print(raw.info)
```

2、 绘制原始数据波形图

```python
raw.plot(duration = 5, n_channels = 25, clipping = None)
```

3、 绘制原始数据功率谱图

```python
raw.plot_psd(average = True)

"""
通过功率谱图看到哪个Hz处可能存在环境噪音
ps：在中国大陆及港澳地区（除台湾省）采集的数据一般工频会出现在50Hz处
"""
```

4、 数据预处理

```python
events, _ = mne.events_from_annotations(raw)

raw.load_data()

raw.filter(7., 35., fir_design='firwin')

raw.info['bads'] += ['EOG-left', 'EOG-right', 'EOG-central']

picks = mne.pick_types(raw.info, meg=False, eeg=True, eog=False, stim=False, exclude='bads')

# 数据切断
tmin, tmax = 2, 6

# left_hand = 769, right_hand = 770, foot = 771, tougue = 772
event_id = dict({'769': 7, '770': 8, '771': 9, '772': 10 })

epochs = mne.Epochs(raw, events, event_id, tmin, tmax, proj=True, picks=picks, baseline=None, preload=True)
```

```python
print(epochs)
```

```python
epochs_data = epochs.get_data()
print(epochs_data.shape)
print(epochs_data)
type(epochs_data)
```

```python
data = epochs_data[:,:,:-1]
print(data.shape)
```

```python
import numpy as np
# 保存为npy的形式
BCI_IV_2a_data = np.save('A01T.npy', data)
```
